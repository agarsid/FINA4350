```
import pickle
f = open('speech.pickle', 'rb')
mydict = pickle.load(f)
f.close()
```

```
chairmandict = mydict[mydict['speaker'].str.startswith('Chairman')] 
```

```
vicechairmandict = mydict[mydict['speaker'].str.startswith('Vice Chair')] 
```

```
governordict = mydict[mydict['speaker'].str.startswith('Governor')] 
```

```
with open('./chairman.pickle', "wb") as output_file:
  pickle.dump(chairmandict, output_file)
```

```
with open('./vicechair.pickle', "wb") as output_file:
  pickle.dump(vicechairmandict, output_file)
```

```
with open('./governor.pickle', "wb") as output_file:
  pickle.dump(governordict, output_file)
```
